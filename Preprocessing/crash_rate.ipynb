```python
import pandas as pd
import numpy as np
from opencage.geocoder import OpenCageGeocode
from tqdm import tqdm
```

```python
traffic_df = pd.read_csv("../data/traffic_counts.csv")
traffic_df = traffic_df.drop(['Unnamed: 2', 'Unnamed: 3'], axis=1)
```

```python
traffic_df.columns
```

```python
street_names = traffic_df.street.unique()
for name in street_names:
    print(name)
```

```python
len(street_names)
```

```python
traffic_df.isna().sum()
```

```python
traffic_df = traffic_df.dropna(subset=['Total'])
```

```python
total_vol = pd.DataFrame(traffic_df.groupby(by=['street'])['Total'].sum(numeric_only=False))
total_vol.reset_index(inplace=True)
```

```python
display(total_vol)
```

```python
total_vol.to_csv('total_vol.csv')
```

```python
df = pd.read_csv(r'crash_ratev2.csv')
df.drop(columns='Unnamed: 0', inplace=True)

# Apply LA at the end. Some streets pointed to in London! 
df['street'] = df['street'].apply(lambda x: x+', Los Angeles')
addresses = df["street"].tolist()

# Go to opencageAPI in google, make an account and then retrieve an API key
key = '4dda65f25f7049319ff93133a38c47c4' # add key here
geocoder = OpenCageGeocode(key)
latitudes = []
longitudes = []
for index, address in tqdm(enumerate(addresses)): 
    result = geocoder.geocode(address, no_annotations="1")  
    
    if result and len(result):  
        longitude = result[0]["geometry"]["lng"]  
        latitude = result[0]["geometry"]["lat"] 
    else:  
        longitude = "N/A"  
        latitude = "N/A"  
    
    latitudes.append(latitude) 
    longitudes.append(longitude)

df['latitude'] = latitudes
df['longitude'] = longitudes
```
